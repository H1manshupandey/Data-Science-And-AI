```python
def compound_interest(principle, rate, years):
    
    comp_interest=0
    
    for i in range (1,years + 1):
        
        simple_interest = (principle * rate * 1)/100
        
        comp_interest = comp_interest + simple_interest
        
        principle=principle + comp_interest
        
    return comp_interest
```

```python
    def romanToInt(s):
    
        # rti is a dict for roman to intgers values
        rti = {'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000}

        # ans is for our sum value
        ans=0

        # for loop till len(s)-1 cause for last Roman Value we cant compare
        for i in range(len(s)-1):
            if rti[s[i]] < rti[s[i+1]]:
                ans = ans - rti[s[i]]
                print(rti[s[i]], " ------  ", ans)
            else:   
                ans = ans + rti[s[i]]
                print(rti[s[i]], " ------  ", ans)

        # So we add the last value and return the final ans
        return ans+rti[s[-1]]
```

```python
def dict_tuple_func(list_of_integers):

    sum = 0

    d = dict()

    for integer in list_of_integers:

        if integer in d:

            d[integer] = d[integer] + 1

        else:

            d[integer] = 1

            sum = sum + integer

    return (d, sum)
```

```python
x = 'My 2 favoriate numbers are 198769.5 And 4565x5'
y = re.findall('[0-9]+.[0-9]+', x) #+ means one or more characters
print(y)
```

```python
x = 'My 2 favoriate numbers are 198769.5 And 45x55'
y = re.findall('[0-9]+\\.[0-9]+', x) #+ means one or more characters
print(y)
```

```python
x = 'My 2 favoriate numbers,  are / "198769.5" And "4565x5"'
y = re.findall('[^0-9a-zA-z]', x) #+ means one or more characters
print(y)
```

```python
x = 'My email address is himanshuxtera@gmail.com'
y = re.findall('\S+@\S+', x) #+ means one or more characters
print(y)
```

```python
## Warning Greedy
x = 'From: Using the : character'
y = re.findall('^F.+:', x) #+ means one or more characters
print(y)
```

```python
## Non Greedy
x = 'From: Using the : character'
y = re.findall('^F.+?:', x) #+ means one or more characters
print(y)
```

```python
## Non Greedy
x = 'From: Using the : character'
y = re.findall('^F[^ ]+:', x) #+ means one or more characters
print(y)
```

```python
import socket
import sys

hostname = "www.bing.com"

try:
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    print ("Socket successfully created here")
except socket.error as err:
    print ("Socket creation failed here with error %s" %(err))

# default port for socket
port = 80

try:
    host_ip = socket.gethostbyname(hostname)
except socket.gaierror:
    print ("Error resolving the host")
    sys.exit()

# connecting to the server
s.connect((host_ip, port))

print (f"Socket successfully connected to {hostname}")
```

```python
# Filter Function

foo = [2,18,9,22,17,24,8,12,27]
divisible_by_3 = filter(lambda x: x%3 ==0, foo)
print (list(divisible_by_3))
```

```python
def myfilter(func, list):
    newlist = []
    for e in list:
        if func(e):
            newlist.append(e)
    return newlist
```

```python
print(myfilter(lambda x: x%3 ==0, foo))
```

```python
def is_div_by3(x):
    return x%3 ==0
```

```python
print(myfilter(is_div_by3, foo))
```
